### 2. 함수로 만들기

```python
def delivery_reply(pageno):
    
    import urllib.request
    from bs4 import BeautifulSoup
    import numpy as np
    import pandas as pd

    pageno = pageno
    url = "http://kin.naver.com/search/list.nhn?query=%EC%B7%A8%EB%93%9D%EC%84%B8&dirId=40306&section=qna&page=" + str(pageno)
    f = urllib.request.urlopen(url)
    overlog = f.read() .decode("utf8")
    f.close()
    soup = BeautifulSoup(overlog, "lxml")
    
    date = list(lv.get_text().replace(".","") for lv in soup.find_all('dd', {'class': 'txt_inline'}))
    title = list(lv.get_text() for lv in soup.find_all('a', {'class': '_nclicks:qna.txt'}))
    category = list(lv.get_text() for lv in soup.find_all('a', {'class': 'txt_g1 _nclicks:qna.cat2'}))
    reply = list(lv.get_text().replace("답변수 ","") for lv in soup.find_all('span', {'class': 'hit'}))
    vote = list(lv.get_text().split('추천수 ')[1].split('\n')[0] for lv in soup.find_all('dd', {'class': 'txt_block'}))
    href = list(str(lv).split("href=")[1].split("tabindex")[0].split(">")[0].replace('"', "").replace(" ", "").replace("amp;","").replace("'","") for lv in soup.find_all('a', {'class': "_nclicks:qna.txt"})[0:])

    delivery = pd.DataFrame({'date':date, 'title':title, 'category':category, 'href':href, 'vote':vote, 'reply':reply})
    delivery = delivery[['date', 'title','category', 'reply', 'vote', 'href']]
    
    return (delivery)
```

```python
delivery_reply(1)
```

```python
delivery_reply(2)
```

### 3. for문 돌려서 여러개 끌어오기

```python
dataset=[]
for i in range(5):
    dataset.append(delivery_reply(i))
```

```python
#pd.set_option('max_colwidth',200)
pd.set_option('max_colwidth',60)
```

```python
fulldata = pd.concat(dataset)
fulldata
```

```python
def delivery_reply2(pageno2):
    
    delivery_dataset=[]
    
    for i in range(pageno2):
        delivery_dataset.append(delivery_reply(i))
    
    delivery_fulldata = pd.concat(delivery_dataset)
    delivery_fulldata.reset_index()
    delivery_fulldata.to_csv('delivery_fulldata.csv')
    
    return (fulldata)
```
